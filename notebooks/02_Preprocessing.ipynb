```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
project_path = "/content/drive/MyDrive/Network_Intrusion_Detection_Project/data/"

train = pd.read_csv(project_path + "UNSW_NB15_training-set.csv")
test = pd.read_csv(project_path + "UNSW_NB15_testing-set.csv")

print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
train.drop(columns=['id'], inplace=True)
test.drop(columns=['id'], inplace=True)
```

```python
print(train.isnull().sum().sum())
print(test.isnull().sum().sum())
```

```python
train.fillna(0, inplace=True)
test.fillna(0, inplace=True)
```

```python
train.fillna(0, inplace=True)
test.fillna(0, inplace=True)
```

```python
y_train = train['label']
y_test = test['label']

X_train = train.drop(['label'], axis=1)
X_test = test.drop(['label'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# Detect categorical columns from BOTH datasets
categorical_cols = list(set(
    X_train.select_dtypes(include='object').columns.tolist() +
    X_test.select_dtypes(include='object').columns.tolist()
))

print("Categorical columns:", categorical_cols)

for col in categorical_cols:
    le = LabelEncoder()

    # 🔥 Force both to string to avoid mixed types
    X_train[col] = X_train[col].astype(str)
    X_test[col] = X_test[col].astype(str)

    combined = pd.concat([X_train[col], X_test[col]], axis=0)

    le.fit(combined)

    X_train[col] = le.transform(X_train[col])
    X_test[col] = le.transform(X_test[col])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaling successful ✅")
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_scaled, y_train = smote.fit_resample(X_train_scaled, y_train)

print("After SMOTE:", y_train.value_counts())
```

```python
X_train_scaled.to_csv(project_path + "X_train_processed.csv", index=False)
X_test_scaled.to_csv(project_path + "X_test_processed.csv", index=False)

y_train.to_csv(project_path + "y_train.csv", index=False)
y_test.to_csv(project_path + "y_test.csv", index=False)
```
